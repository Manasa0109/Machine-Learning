```python
#copying arrays
a=np.array([1,2,3])
b=a
b[0]=100
print(b)#same reference
print(a)#same reference
```

statistics


```python
stats=np.array([[1,2,3],[4,5,6]])
print(np.min(stats,axis=1))#gives min in rows
print(np.min(stats,axis=0))#gives min in columns

```

```python
print(np.max(stats,axis=1))#gives max in rows
print(np.max(stats,axis=0))#gives max in columns
```

```python
np.sum(stats)
```

reorganinsing arrays

```python
before=np.array([[1,2,3,4],[5,6,7,8]])
print(before)
```

```python
after=before.reshape((2,2,2))
print(after)
```

```python
#verticlly stacking vectors
v1=np.array([1,2,3,4])
v2=np.array([5,6,7,8])
print(np.vstack([v1,v2]))
print(np.vstack([v1,v2,v1,v2]))
```

```python
#horizontal stack
h1=np.ones((2,4))
h2=np.zeros((2,2))
np.hstack((h1,h2))
```
